```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split 
from sklearn.metrics import classification_report
```

```python
df = pd.read_csv("emotions.csv")
```

```python
df
```

```python
df.head()
```

```python
df['# mean_0_a'].dtype
columns = df.loc[:, '# mean_0_a' : 'fft_749_b']
```

```python
columns
```

```python
#Setting Positive and Neutral as 0 and Negative as 1
df['label'] = df['label'].replace(['NEGATIVE', 'NEUTRAL', 'POSITIVE'], [0,0,1])
```

```python
df.label
```

<h1>Logistic Regression</h1>

```python
reg = linear_model.LogisticRegression()
X_Test, X_Train, Y_Test, Y_Train = train_test_split(columns, df.label, test_size = 0.8)
```

```python
reg.fit(X_Train, Y_Train)
```

```python
reg.score(X_Test,  Y_Test)
print(reg.score(X_Test, Y_Test))
```

```python
reg.score(X_Train, Y_Train)
```

```python
y_pred = reg.predict(X_Test)
```

```python
y_pred = reg.predict(X_Test)
print(classification_report(Y_Test, y_pred))
```

```python
reg = linear_model.LogisticRegression()
X_Test, X_Train, Y_Test, Y_Train = train_test_split(columns, df.label, test_size = 0.2)
reg.fit(X_Train, Y_Train)
print('Accuracy obtained =', reg.score(X_Test,  Y_Test) * 100)
print(len(X_Train))
print(len(X_Test))
y_pred = reg.predict(X_Test)
y2_pred = reg.predict(X_Train)
print("Report for Testing Set")
print(classification_report(Y_Test, y_pred))
print("For Training Set : ")
print(classification_report(Y_Train, y2_pred))
```

```python
reg = linear_model.LogisticRegression()
X_Test, X_Train, Y_Test, Y_Train = train_test_split(columns, df.label, test_size = 0.3)
reg.fit(X_Train, Y_Train)
print('Accuracy obtained =', reg.score(X_Test,  Y_Test) * 100)
print(len(X_Train))
print(len(X_Test))
y_pred = reg.predict(X_Test)
y2_pred = reg.predict(X_Train)
print("Report for Testing Set")
print(classification_report(Y_Test, y_pred))
print("For Training Set : ")
print(classification_report(Y_Train, y2_pred))
```
